```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
sns.set_style('whitegrid')
```

```python
simulations_names = [f'S{i}' for i in range(1,7) if i != 2]
simulations_sizes = ['35k', '100k', '300k', '600k', '1M']
best_sample_sizes_perc = [1, 0.05, 0.05, 0.1, 0.05]
best_sample_sizes_times = [11.68, 56.35, 108.82, 223.16, 355.63]
simulations_names = [x + f' ({y})' for x,y in zip(simulations_names, simulations_sizes)]
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12,4))
axes = axes.flatten()  

sns.lineplot(x=simulations_names, y=best_sample_sizes_perc, color='blue', marker='o',  markersize=5, ax=axes[0])
sns.lineplot(x=simulations_names, y=best_sample_sizes_times, color='blue', marker='o', markersize=5, ax=axes[1])

axes[0].set_title('Relative Sample Size vs Simulations', size=13, weight='bold')
axes[0].set_ylabel('Relative Sample Size (%)', size=11)
axes[0].set_yticks(np.round(np.linspace(np.min(best_sample_sizes_perc), np.max(best_sample_sizes_perc), 6),3))
axes[0].axhline(y=np.mean(best_sample_sizes_perc), color='red', linestyle='--', label='Average')
axes[0].legend()

axes[1].set_title('Time vs Simulations', size=13, weight='bold')
axes[1].set_ylabel('Time (secs)', size=11)
axes[1].set_yticks(np.round(np.linspace(np.min(best_sample_sizes_times), np.max(best_sample_sizes_times), 6),0))

for i in range(len(axes)):
    axes[i].set_xlabel('Simulations', size=11)


#plt.tight_layout()
plt.subplots_adjust(top=0.83)
plt.suptitle('Best sample size for each simulation\nFast $k$-Medoids', fontsize=14, y=1.03, weight='bold', color='black', alpha=1)
plt.subplots_adjust(hspace=0.5, wspace=0.23) 

file_name = 'plots/kmedoids_simulation_best_sample_sizes'
fig.savefig(file_name + '.jpg', format='jpg', dpi=500, bbox_inches="tight", pad_inches=0.2)

plt.show()
```
